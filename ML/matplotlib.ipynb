```python
import matplotlib.pyplot as plt
import numpy as np

# Line Chart Example
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]
plt.plot(x, y, color='blue', linestyle='-', marker='o', linewidth=2, markersize=6)
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Bar Chart Example
x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]
plt.bar(x, y, width=0.5, color='green', edgecolor='black')
plt.title("Bar Chart Example")
plt.show()
```

```python
# Horizontal Bar Chart
x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]
plt.barh(x, y, color='orange')
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
# Multiple Bar Graph
x = np.arange(4)
y1 = [3, 7, 2, 5]
y2 = [4, 6, 3, 4]
bar_width = 0.4
plt.bar(x, y1, width=bar_width, color='blue', label='Group 1')
plt.bar(x + bar_width, y2, width=bar_width, color='red', label='Group 2')
plt.xticks(x + bar_width / 2, ['A', 'B', 'C', 'D'])
plt.legend()
plt.title("Multiple Bar Graph")
plt.show()
```

```python
# Pie Chart
slices = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']
explode = [0, 0.2, 0.3, 0]
plt.pie(slices, labels=labels, autopct='%.1f%%', shadow=True)
plt.title("Pie Chart")
plt.show()
```

```python
# Histogram
data = [12, 15, 22, 20, 18, 30, 16, 25, 10, 10, 28, 22]
plt.hist(data, bins=[0,10,20,30], rwidth=0.6, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Frequency Polygon
data = [12, 15, 22, 20, 18, 30, 16, 25, 10, 10, 28, 22]
plt.hist(data, bins=[0,10,20,30], histtype='step', linewidth=2)
plt.title("Frequency Polygon")
plt.show()
```

```python
# Box Plot
data = [10, 20, 30, 40, 50, 60, 70, 80]
plt.boxplot(data, notch=True, patch_artist=True)
plt.title("Box Plot")
plt.show()
```

```python
# Scatter Plot
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
